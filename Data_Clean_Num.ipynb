```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
content = []
count = 0

determinant_list = [5, 28, 29, 30, 34, 35, 36, 37, 38, 39, 41, 42, 43, 44, 48, 51, 52, 53]

time_start = time.time()

with open('gallup_all_compileCol.txt') as txtfile:
    for row in txtfile:

        row = row.split(',')
        row[-1] = row[-1].strip()

        if count == 0:
            content.append(row)
        else:
            content.append(row[1:])

        count += 1

content_mat = np.array(content)
num_sample, num_var = content_mat.shape

time_end = time.time()
print('Reading data is complete! Running time is ' + str(time_end - time_start) + 's!')


discrete_list = [0, 1, 2, 3, 4]
float_list = [12, 13, 14, 15, 16, 17, 21, 40, 59, 60, 61, 62]


def format_conv(mat, i):
    for j in range(1, len(mat[1:, i]) + 1):
        if mat[j, i] == '""' or mat[j, i] == 'NA':
            mat[j, i] = -10
        else:
            mat[j, i] = mat[j, i].strip('"')
            mat[j, i] = mat[j, i].strip('+')

    print("iteration " + str(i) + " is complete!")


for i in range (0, num_var):
    # state_of_econ cleaning
    if i == 0:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'Excellent':
                    content_mat[j, i] = 4
                elif content_mat[j, i] == 'Good':
                    content_mat[j, i] = 3
                elif content_mat[j, i] == 'Only fair':
                    content_mat[j, i] = 2
                elif content_mat[j, i] == 'Poor':
                    content_mat[j, i] = 1
        print("iteration " + str(i) + " is complete!")

    # satis_future cleaning
    elif i == 1:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'Getting better':
                    content_mat[j, i] = 1
                elif content_mat[j, i] == '(The same)':
                    content_mat[j, i] = 0
                elif content_mat[j, i] == 'Getting worse':
                    content_mat[j, i] = -1
                elif content_mat[j, i] == '(DK)':
                    content_mat[j, i] = 0
                elif content_mat[j, i] == '(Refused)':
                    content_mat[j, i] = 10
        print("iteration " + str(i) + " is complete!")


    #lifeladder_now/future
    elif i == 2 or i == 3:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'Best Possible':
                    content_mat[j, i] = 10
                elif content_mat[j, i] == 'Worst Possible':
                    content_mat[j, i] = 0

        print("iteration " + str(i) + " is complete!")


    #healthprob_days:
    elif i == 4:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'Zero days':
                    content_mat[j, i] = 0

        print("iteration " + str(i) + " is complete!")


    # time_commute & time_social:
    elif i == 5 or i == 6:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == '(DK/Does not apply/Work from home)':
                    content_mat[j, i] = 0
                elif content_mat[j, i] == 'None':
                    content_mat[j, i] = 0
                elif content_mat[j, i] == 'Less than one hour':
                    content_mat[j, i] = 0.5
                elif content_mat[j, i] == '(DK)':
                    content_mat[j, i] = -2

        print("iteration " + str(i) + " is complete!")


    #polview
    elif i == 8:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'Very Conservative':
                    content_mat[j, i] = -2
                elif content_mat[j, i] == 'Conservative':
                    content_mat[j, i] = -1
                elif content_mat[j, i] == 'Moderate':
                    content_mat[j, i] = 0
                elif content_mat[j, i] == 'Liberal':
                    content_mat[j, i] = 1
                elif content_mat[j, i] == 'Very Liberal':
                    content_mat[j, i] = 2

        print("iteration " + str(i) + " is complete!")


    #nchild
    elif i == 9:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'None':
                    content_mat[j, i] = 0

        print("iteration " + str(i) + " is complete!")


    #education
    elif i == 64:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'educ_nohs':
                    content_mat[j, i] = 1
                elif content_mat[j, i] == 'educ_hs':
                    content_mat[j, i] = 2
                elif content_mat[j, i] == 'educ_tech':
                    content_mat[j, i] = 3
                elif content_mat[j, i] == 'somecoll':
                    content_mat[j, i] = 4
                elif content_mat[j, i] == 'edu_coll':
                    content_mat[j, i] = 5
                elif content_mat[j, i] == 'educ_post':
                    content_mat[j, i] = 6

        print("iteration " + str(i) + " is complete!")


    #race
    elif i == 65:
        for j in range(1, len(content_mat[1:, i]) + 1):
            if content_mat[j, i] == '""' or content_mat[j, i] == 'NA':
                content_mat[j, i] = -10
            else:
                content_mat[j, i] = content_mat[j, i].strip('"')
                content_mat[j, i] = content_mat[j, i].strip('+')

                if content_mat[j, i] == 'white':
                    content_mat[j, i] = 1
                elif content_mat[j, i] == 'black':
                    content_mat[j, i] = 2
                elif content_mat[j, i] == 'hisp':
                    content_mat[j, i] = 3
                elif content_mat[j, i] == 'asian':
                    content_mat[j, i] = 4

        print("iteration " + str(i) + " is complete!")

    else:
        format_conv(content_mat, i)

content_mat = np.delete(content_mat, determinant_list, axis= 1)

filetoWrite = open("gallup_clean_nocol8.txt", "w+")
for line in content_mat:
    filetoWrite.write(",".join(line) + "\n")

filetoWrite.close()
```
