```python
import time
import numpy as np
import random
```

```python
def parseFile(file):
    time_start = time.time()

    content = []
    count, count_incomplete,count_complete, count_part = 0, 0, 0, 0

    with open(file) as txtfile:
        for row in txtfile:

            row = row.split(',')
            row[-1] = row[-1].strip()

            content.append(row)

            count += 1

    content_mat = np.array(content)

    time_end = time.time()
    print('Reading data is complete! Running time is ' + str(time_end - time_start) + 's!')

    return content_mat
```

```python
def fill_mean_int(np_col, missing_val):

    col = np_col[1:]
    # count entries in each col that aren't None
    count = np.count_nonzero(col != missing_val)
    # find index of 'No Entry' in each col
    indices = col  == missing_val
    # replace Missing values with 0
    col[indices] = 0

    col = col.astype(np.int)
    # compute mean of each col
    mean = np.sum(col)/ count
    # replace None in each col with the mean of that col
    col[indices] = int(mean)

    return col

def fill_mean_float(np_col, missing_val):
    col = np_col[1:]
    # count entries in each col that aren't None
    count = np.count_nonzero(col != missing_val)
    # find index of 'No Entry' in each col
    indices = col == missing_val
    # replace Missing values with 0
    col[indices] = 0

    col = col.astype(np.float)
    # compute mean of each col
    mean = np.sum(col) / count
    # replace None in each col with the mean of that col
    col[indices] = mean

    return col


def fill_mean_time_pass(np_col, missing_val):
    # don't do anything for string
    return np_col[1:]
```

```python
def fill_random_sample_int(np_col, missing_val):
    col = np_col[1:]
    # count entries in each col that aren't None
    count = np.count_nonzero(col != missing_val)
    # find index of 'No Entry' in each col
    indices = np.where(col == missing_val)

    sample_indices = np.where(col != missing_val)

    for idx in indices[0]:
        rand_samp = random.randint(0, count-1)
        temp_idx = sample_indices[0][rand_samp]
        col[idx] = col[temp_idx]

    col = col.astype(np.int)

    return col
```

```python
def writeToFile(mat, writeToFilePath):
    filetoWrite = open(writeToFilePath, "w+")
    for line in mat:
        filetoWrite.write(",".join(line) + "\n")

    filetoWrite.close()
```

```python
def main():
    path = 'gallup_clean_NA_determinant.txt'
    content_mat = parseFile(path)

    int_var_list = [0, 1, 2, 3, 4, 6, 7, 8, 9, 21, 22, 23, 24, 25, 26, 27, 28, 29, 31, 32, 33, 34, 35, 36, 37, 38, 45, 46, 47]
    float_var_list = [5, 11, 15, 20, 30, 39, 40, 40, 41, 42, 43, 44]

    output_array = []

    num_sample, num_var = content_mat.shape

    for i in range(0, num_var):
        if i in int_var_list:
            temp_out_put = fill_mean_int(content_mat[:, i], 'N/A') # mean filling
            #temp_out_put = fill_random_sample_int(content_mat[:, i], 'N/A') # random sampling
            output_array.append(temp_out_put)
            print('Column ' + str(i) + ' has been filled with mean value')
        elif i in float_var_list:
            temp_out_put = fill_mean_float(content_mat[:, i], 'N/A')
            output_array.append(temp_out_put)
            print('Column ' + str(i) + ' has been filled with mean value')

        else:
            temp_out_put = fill_mean_time_pass(content_mat[:, i], 'N/A')
            output_array.append(temp_out_put)
            print('Column ' + str(i) + ' has been filled with mean value')

    output_array = np.array(output_array).astype(np.str).transpose()

    writeToFile(output_array, 'gallup_mean_filled_cleaned.txt')


if __name__ == "__main__":
    main()
```
